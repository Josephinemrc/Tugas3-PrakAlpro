```python
try:
  a = int(input("masukkan bilangan pertama: "))
  b = int(input("masukkan bilangan kedua: "))
  c = int(input("masukkan bilangan ketiga: "))

  if a > b and a > c:
    print("terbesar:", a)
  elif b > a and b > c:
    print("terbesar", b)
  elif c > a and c > b:
    print("terbesar:", c)
except:
  print("Format yang anda masukkan salah")
```

```python
try:
  sisi_satu = float(input("masukkan sisi pertama: "))
  sisi_dua = float(input("masukkan sisi keduaa: "))
  sisi_tiga = float(input("masukkan sisi ketiga: "))
  if sisi_satu == sisi_dua == sisi_tiga:
    print("3 sisi sama")
  elif sisi_satu == sisi_tiga != sisi_dua or sisi_satu == sisi_dua != sisi_tiga or sisi_dua == sisi_tiga != sisi_satu:
    print("2 sisi sama")
  else:
    print("tidak ada yang sama")
except:
  print("Format yang anda masukkan salah")
```
